# Load dataset

```python
import pandas as pd

file_path = r"C:\Users\Noureen\Desktop\archive(11)\summer-products-with-rating-and-performance_2020-08.csv"
summer_products = pd.read_csv(file_path)

print(summer_products.head())
```

```python
# Print the column names
print(summer_products.columns.tolist())
```

# Splitting data into dimension

```python
product_dimension = summer_products[
    ['product_id', 'title', 'title_orig', 'price', 'retail_price', 
     'currency_buyer', 'tags', 'product_color', 'product_variation_size_id', 
     'product_variation_inventory', 'inventory_total', 'product_url', 'product_picture', 'theme']
].copy()

product_dimension
```

```python
# Merchant Dimension Table
merchant_dimension = summer_products[
    ['merchant_id', 'merchant_title', 'merchant_name', 'merchant_info_subtitle', 
     'merchant_rating_count', 'merchant_rating', 'merchant_has_profile_picture', 
     'merchant_profile_picture']
].copy()

merchant_dimension 
```

```python
# Shipping Dimension
shipping_dimension = summer_products[
    ['product_id', 'shipping_option_name', 'shipping_option_price', 
     'shipping_is_express', 'countries_shipped_to', 'origin_country']
].copy()

shipping_dimension
```

```python
# Rating Dimension
rating_dimension = summer_products[
    ['product_id', 'rating', 'rating_count', 'rating_five_count', 'rating_four_count', 
     'rating_three_count', 'rating_two_count', 'rating_one_count']
].copy()

rating_dimension
```

# Transforming Date Dimension

```python
date_dimension = summer_products[["crawl_month"]]
date_dimension
```

```python
date_dimension['crawl_month']=pd.to_datetime(date_dimension['crawl_month'])


date_dimension['day'] = date_dimension['crawl_month'].dt.day
date_dimension['month'] = date_dimension['crawl_month'].dt.month
date_dimension['year'] = date_dimension['crawl_month'].dt.year
```

```python
date_dimension["date_id"]=range(1,len(date_dimension)+1)

date_dimension
```

# Data Cleaning

```python
# Drop duplicates
product_dimension = product_dimension.drop_duplicates(subset=['product_id'])
product_dimension = product_dimension.drop_duplicates(subset=['product_picture'])
product_dimension = product_dimension.drop_duplicates(subset=['product_url'])
merchant_dimension = merchant_dimension.drop_duplicates(subset=['merchant_id'])  
date_dimension = date_dimension.drop_duplicates(subset=['date_id'])
rating_dimension = rating_dimension.drop_duplicates()
shipping_dimension['shipping_id'] = shipping_dimension.index + 1
```
